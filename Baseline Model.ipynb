### Multicolinearity

Multicollinearity occurs when two or more predictors are highly correlated to each other, such that they do not provide unique or independent information in the regression model. ### Modeling

Now, let's model our baseline model!

In variable selection, I will use the **p-value threshold of 0.5** to determine statistical significance of features.

```python
import statsmodels.api as sm
```

I am going to make a function that would instantiate, fit, summarize, and calculate the model accuracy all at once given the target and feature variables.

For the baseline model, I will set default paramaters for the logistic regression model:
- No regularization (C=1)
- Solver = lbfgs

```python
def model_summary(x, y):
    # adding the constant
    X_const = sm.add_constant(X)

    # instantiating the model
    logit = sm.Logit(y, X_const)

    # fitting the model to the data
    logit_fitted = logit.fit()

    # looking at the results 
    summary = logit_fitted.summary()

    # Calculate soft predictions
    y_proba = logit_fitted.predict(X_const)

    # Convert soft predictions to hard predictions 0/1
    y_pred = np.where(y_proba >= 0.5, 1, 0)

    # Calculate # correct
    num_correct = (y_pred == y).sum()

    # Calculate the percentage accuracy
    pct_accuracy = num_correct / X.shape[0]
    
    print(f'\nThe model accuracy is {round((pct_accuracy*100.0),2)}%')
    
    return summary
```

### Model Evaluation

With class imbalance, the cost function is going to be biased towards the majority class (failed delivery) and will result in a model that doesn't learn enough about the minority class (successful delivery). Since we have an imbalanced data, we should also rely on other evaluation metrics of the model than accuracy: precision, recall, F1 score, and AUC.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #for quick modeling
from sklearn.metrics import classification_report
```

```python
#the chosen model: Model 1.0
X = df[["Shipping Mode_Standard Class"]]
y = df["Delivery Status"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y, random_state=123)
```

I split the model into 70-30 train-test split.

```python
# Instantiate model
logit = LogisticRegression(solver='lbfgs', random_state=123, max_iter=10000)

# Fit and predict model
logit.fit(X_train, y_train)

# Model Accuracies
train_acc = logit.score(X_train, y_train)
test_acc = logit.score(X_test, y_test)
print(f"Train Accuracy: {train_acc}")
print(f"Test Accuracy: {test_acc}")
```

We expected an accuracy of around 68.27%, and we get accuracies around this number here.

We can see that the model is not overfitting at all, as the test accuracy is even higher than the train accuracy. Our test accuracy, nonetheless, is higher than our expectation: a **test set accuracy of 68.43%**!

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Get class predictions
y_pred = logit.predict(X_test)

# Generate confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)

# label rows and columns
cf_df = pd.DataFrame(
    cf_matrix, 
    columns=["Predicted Failed Delivery", "Predicted Successful Delivery"],
    index=["True Failed Delivery", "True Successful Delivery"]
)

display(cf_df)
```

```python
# the function expects the estimator, inputs and target as parameters
ConfusionMatrixDisplay.from_estimator(logit, X_test, y_test, values_format='d')
```

```python
print("Classification Report for TEST dataset\n")
print(classification_report(y_test, y_pred))
```

Here we see that the model tends to make a lot of **False Positives**, hence low precision score for predicting "1". This means that the model would tend to identify order items to be succesfully delivered, when it would actually be a failure.

On the other hand, the model has a few **False Negatives**, hence relatively high recall score for predicting "1". This means that the model make few mistakes in predicting successful delivereies as a failed one.

Let's plot out an ROC curve and calculate AUC:

```python
#define our own false positive rate calculation, as sklearn does not have this functionality
def false_positive_rate(y_true, y_pred):
    
    false_positives = (y_true == 0) & (y_pred == 1) # find all values where y is negative
                                                    # but we predicted positive
    false_positive_number = false_positives.sum()

    true_negatives = (y_true == 0) & (y_pred == 0)  # find all values where y is negative
                                                    # and we predicted negative
    true_negative_number = true_negatives.sum()

    # Finally, find the ratio of (FP) to (TN + FP)
    FPR = false_positive_number/(true_negative_number + false_positive_number)
    
    return FPR
```

```python
from sklearn.metrics import recall_score

#Get the probability for each point in the test set.
logit = LogisticRegression()
y_proba = logit.fit(X_train, y_train).predict_proba(X_test)[:,1]
thresholds = np.linspace(0, 1, 100)

# Iterate over the thresholds and compute the positive and negative recall
true_positive_rates = []
false_positive_rates = []

thresholds = np.linspace(0, 1, 50)
for threshold in thresholds:
    
    # Apply the cutoff and calculate the positive and negative recall
    y_pred = np.where(y_proba > threshold, 1, 0)
    
    tpr = recall_score(y_test, y_pred)
    fpr = false_positive_rate(y_test, y_pred)
    
    true_positive_rates.append(tpr)
    false_positive_rates.append(fpr)

# True positive rate (TPR) = recall
tprs = np.array(true_positive_rates)

# False positive rate (FPR)
fprs = np.array(false_positive_rates)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

#Get the probability for each point in the train set.
y_proba_train = logit.predict_proba(X_train)[:,1]

# Compute ROC curve and AUC for for the one class
fprs_train, tprs_train, thresholds_train = roc_curve(y_train, y_proba_train)
roc_auc_train = roc_auc_score(y_train, y_proba_train)
roc_auc_test = roc_auc_score(y_test, y_proba)
  
# Plot the ROC curve.
plt.figure()
plt.plot(fprs_train, tprs_train, color='darkorange', lw=2, label='train')
plt.plot(fprs, tprs, lw=2, label='test')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC and AUC')
plt.legend(loc="best")
plt.show()
print(f"Train AUC score: {roc_auc_train}")
print(f"Test AUC score: {roc_auc_test}")
```

In our case, we want a model that is **more stringent on predicting successful deliveries** meaning that we want the model to predict successful deliveries only if it has very high confidence. This is because **the cost of a false positive prediction outweighs the cost of making a false negative prediction**. The cost of misclassifying order items that will fail its delivery is the potential revenue it could have generated or the extra customer satisfaction that could otherwise be gained. On the other hand, the cost of misclassifying order items that will succesfully be delivered is the extra time and effort made to ensure a successful delivery. We can see that there is **less risk associated with making false positive predictions than false negative predictions**. 

This means that we desire a model that performs with low False Positive Rate (FPR).

Our baseline model, however, has high instances false positives. Therefore, the baseline model poorly satisfies our desired model performance of having low FPR.

The model has a test AUC score of 0.7085. Based on this, we know that given two random order items, one from the successful delivery class and one from the failed delivery class, our model is **70.85% confident that the positive point belongs to the positive class than the negative point belonging to the positive class**.

Since we want a model that performs with low FPR, we will want a higher AUC score so that the model is more strict at classifying a successful delivery.

A potential reason for the baseline's **high instance of false positives is due to the class imbalance** where there are more data on failed deliveries (0) than that of successful deliveries (1).
